```python
from qjm import Wargame

wargame = Wargame()
wargame.load_sim_state('./wargames/saves/scenario_save.sav')

desired_formation = '5/6'
for form_id in wargame.formationsById:
    form = wargame.formationsById[form_id]
    if form.fullshortname == desired_formation:
        print(form)
        print(form.get_oli())

desired_formation = '4/6'
for form_id in wargame.formationsById:
    form = wargame.formationsById[form_id]
    if form.fullshortname == desired_formation:
        print(form)
        print(form.get_oli())
```

```python
# Get all QJM equipment types and print their OLI values
qjm_equip = []
for name in wargame.equipment_database.vehicles:
    qjm_equip.append((name, wargame.equipment_database.vehicles[name].q_OLI))
for name in wargame.equipment_database.weapons:
    qjm_equip.append((name, wargame.equipment_database.weapons[name].q_OLI))

# Sort equipment by OLI value
qjm_equip.sort(key=lambda x: -x[1])
for eq, oli in qjm_equip:
    print(f'{eq:>20s}: {oli:8,.2f}')
```

```python
# Look at 9K32 Strela-2 to figure out why its value is negative
equip = wargame.equipment_database.weapons['9K32 Strela-2']
GLOBAL_DISPERSION = 4000

q_OLI = (equip.q_RF * equip.q_PTS * equip.q_RIE * equip.q_RN *
            equip.q_A * equip.q_RL * equip.q_SME * equip.q_MBE *
            equip.q_MCE * equip.q_GE * 
            equip.q_MRN * equip.q_PEN * equip.q_VEL *
            equip.d_ENa * equip.d_ENo / GLOBAL_DISPERSION)

print(f"{equip.name}: {equip.q_RF} * {equip.q_PTS} * {equip.q_RIE} * {equip.q_RN} * \n"
        f"{equip.q_A} * {equip.q_RL} * {equip.q_SME} * {equip.q_MBE} * \n"
        f"{equip.q_MCE} * {equip.q_GE} * {equip.q_MRN} * {equip.q_PEN} * \n"
        f"{equip.q_VEL} * {equip.d_ENa} * {equip.d_ENo} / {GLOBAL_DISPERSION} = {q_OLI:.2f}")
```
